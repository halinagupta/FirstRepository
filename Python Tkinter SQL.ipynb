```python
import mysql.connector as mc
from tkinter import *
conn=mc.connect(host='localhost',user='root',password='MySQL2020',database='mydb')
mycursor=conn.cursor()
root=Tk()
root.title("Employee Management System")
root.geometry("600x400") #You want the size of the app to be 300x300
root.resizable(0, 0) #Don't want the window to be resizble
empid=StringVar()
ename=StringVar()
city=StringVar()
dob=StringVar()
sal=StringVar()
allow=StringVar()
dno=StringVar()
gender=StringVar()
def SearchData():
    
    if(empid.get()!=""):    #This method searches record on the basis of empid provided
        eid=int(empid.get())
        if(eid):
            sql="Select * from emp where empid={}".format(eid)  
            try:
                mycursor.execute(sql)
                records=mycursor.fetchone()
                print(records)
                if(mycursor.rowcount>=1):
                    ename.set(str(records[1]))
                    city.set(str(records[2]))
                    dob.set(str(records[3]))
                    sal.set(str(records[4]))
                    allow.set(str(records[5]))
                    dno.set(str(records[6]))
                    gender.set(str(records[7]))
                else:
                     clearfields()
                     callpopup("Important","Data not found")
                     
                     
            except mysql.connector.errors.InternalError:
                    print("Internal error raised")
    else:
        callpopup("Error","EmpID cannot be empty")
        clearfields()
def UpdateData():
    if(not checkempty()):
        eid=int(empid.get())
        en=ename.get()
        cty=city.get()
        db=dob.get()
        s=int(sal.get())
        a=int(allow.get())   
        dn=int(dno.get())
        gnd=gender.get()
        sql="update emp set ename=%s, city=%s, dob=%s, salary=%s, allowance=%s,deptno=%s, gender=%s where empid=%s"
        val=(en,cty,db,s,a,dn,gnd,eid)
        mycursor.execute(sql,val)
        conn.commit()
        clearfields()
        msg=str(mycursor.rowcount)+" Record updated"
        callpopup("Succesful",msg)
    else:
        callpopup("ERROR","One or more fields are empty")
def callpopup(title,msg): #This method will call out the pop up window with the suitable message
    popup = Tk()
    popup.wm_title(title)
    label=Label(popup, text=msg,font=("Arial Bold", 8))
    label.pack(side="top", fill="x", pady=10)
    B1 = Button(popup, text="Okay", command = popup.destroy)
    B1.pack()
    popup.mainloop()
def clearfields():
    empid.set("")
    ename.set("")
    city.set("")
    dob.set('')
    sal.set("")
    allow.set("")
    dno.set('')
    gender.set('')
def checkempty():
    if(empid.get()=='' or ename.get()=='' or dob.get()=='' or city.get()=='' or allow.get()=='' or dno.get()=='' or gender.get()==''):
        return True
    else:
        return False
def AssignData():
    if(not checkempty()):
        eid=int(empid.get())
        en=ename.get()
        cty=city.get()
        db=dob.get()
        s=int(sal.get())
        a=int(allow.get())   
        dn=int(dno.get())
        gnd=gender.get()
        sql="INSERT INTO emp VALUES (%s, %s, %s, %s, %s,%s,%s,%s)"
        val=(eid,en,cty,db,s,a,dn,gnd)
        mycursor.execute(sql,val)
        conn.commit()
        clearfields()
        msg=str(mycursor.rowcount)+" Record inserted"
        callpopup("Succesful",msg)
    else:
        callpopup("ERROR","One or more fields are empty")
def Makeinterface(): #Tkinter Interface
    label1=Label(root,text='EmpID',font=("Arial Bold", 14))
    label1.grid(row=0,column=0)
    entry1=Entry(root,textvariable=empid)
    entry1.grid(row=0,column=1)
    label2=Label(root,text='EmpName',font=("Arial Bold", 14))
    label2.grid(row=1,column=0)
    entry2=Entry(root,textvariable=ename,width=25)
    entry2.grid(row=1,column=1)
    label3=Label(root,text='City',font=("Arial Bold", 14))
    label3.grid(row=2,column=0)
    entry3=Entry(root,textvariable=city)
    entry3.grid(row=2,column=1)
    label4=Label(root,text='DOB',font=("Arial Bold", 14))
    label4.grid(row=3,column=0)
    entry4=Entry(root,textvariable=dob)
    entry4.grid(row=3,column=1)
    label5=Label(root,text='Salary',font=("Arial Bold", 14))
    label5.grid(row=4,column=0)
    entry5=Entry(root,textvariable=sal)
    entry5.grid(row=4,column=1)
    label6=Label(root,text='Allowance',font=("Arial Bold", 14))
    label6.grid(row=5,column=0)
    entry6=Entry(root,textvariable=allow)
    entry6.grid(row=5,column=1)
    label7=Label(root,text='Department',font=("Arial Bold", 14))
    label7.grid(row=6,column=0)
    entry7=Entry(root,textvariable=dno)
    entry7.grid(row=6,column=1)
    label8=Label(root,text='Gender',font=("Arial Bold", 14))
    label8.grid(row=7,column=0)
    entry8=Entry(root,textvariable=gender)
    entry8.grid(row=7,column=1)
    frmbtn=Frame(root)
    frmbtn.grid(row=8,column=0)
    btnSave=Button(frmbtn,text='Save',command=AssignData)
    #btnSave.grid(row=6,column=0)
    btnSave.pack(side=LEFT)
    btnExit=Button(frmbtn,text='Quit',command=root.destroy)
    #btnExit.grid(row=6,column=1)
    btnExit.pack(side=LEFT)
    btnSearch=Button(frmbtn,text='Search',command=SearchData)
    btnSearch.pack(side=LEFT)
    btnUpdate=Button(frmbtn,text='Update',command=UpdateData)
    btnUpdate.pack(side=LEFT)
    root.mainloop()
Makeinterface()
```
